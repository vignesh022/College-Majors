```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
%matplotlib inline
```

```python
recent_grads = pd.read_csv('recent-grads.csv',encoding='UTF-8')
print(recent_grads.iloc[0])
print('\n')
print(recent_grads.head())
print('\n')
print(recent_grads.tail())
print('\n')
print(recent_grads.describe())
```

```python
raw_data_count = recent_grads.shape[0]

recent_grads = recent_grads.dropna()
cleaned_data_count = recent_grads.shape[0]
```

```python
ax1 = recent_grads.plot(x='Sample_size',y='Median',kind='scatter')
ax1.set_title('Median vs Sample_size')
ax1.set_xlim(-250,5000)

ax2 = recent_grads.plot(x='Sample_size',y='Unemployment_rate',kind='scatter')
ax2.set_title('Unemployment_rate vs Sample_size')
ax2.set_xlim(-250,5000)
ax2.set_ylim(-0.025,0.20)

ax3 = recent_grads.plot(x='Full_time',y='Median',kind='scatter')
ax3.set_title('Median vs Full_time')
ax3.set_xlim(-25000,300000)

ax4 = recent_grads.plot(x='ShareWomen',y='Unemployment_rate',kind='scatter')
ax4.set_title('Unemployment_rate vs ShareWomen')
ax4.set_xlim(-0.1,1.1)
ax4.set_ylim(-0.025,0.20)

ax5 = recent_grads.plot(x='Men',y='Median',kind='scatter')
ax5.set_title('Median vs Men')
ax5.set_xlim(-25000,300000)

ax6 = recent_grads.plot(x='Women',y='Median',kind='scatter')
ax6.set_title('Median vs Women')
ax6.set_xlim(-25000,300000)
```

```python
men_women = recent_grads.groupby("Major_category").agg({"Men": "sum", "Women": "sum"})
men_women.index
```

```python
labels = men_women.index
x = np.arange(16)  # the label locations
width = 0.3 # the width of the bars

fig, ax = plt.subplots(figsize=(10,10))
rects1 = ax.bar(x-width/2, men_women['Men'], width, align='center',label='Men',color="blue")
rects2 = ax.bar(x+width/2, men_women['Women'], width, align='center',label='Women',color="red")

# Add some text for labels, title and custom x-axis tick labels, etc.
ax.set_ylabel('Enrollment')
ax.set_title("Gender based enrollment across the categories of degree majors\n")
ax.set_xticks(x)
ax.set_xticklabels(labels,rotation=90)
ax.legend()
```

```python
fig, ax1 = plt.subplots(figsize=(6,6))
ax1.boxplot(recent_grads["Median"],vert=False)
ax1.set_title("Box-plot of Median earnings of Recent graduates")
fig, ax2 = plt.subplots(figsize=(6,6))
ax2.boxplot(recent_grads["Unemployment_rate"],vert=False)
ax2.set_title("Box-plot of Unemployment Rate of Recent graduates")
```
